```python
#19
catogCount = []
catog = ["WorkClass", "Education", "Marital-Status", "Occupation", "Relationship", "Race", "Sex"]
for i in range(len(catog)):
    print(catogCount.append(df.groupby(catog[i])['Age'].count()))
    
for c in catog:
    oneHot = pd.get_dummies(df[c])
    df = df.drop(c, axis = 1)
    df = df.join(oneHot)    
    
    
```

```python
df.shape
```

```python
#graph
for n, c in zip(catog, catogCount):
    v = c.values
    l = c.index
    vP = [round(i, 2) for i in (v / sum(v) * 100)]
    ex = [0 if i > 4 else 0.2 for i in vP]
    plt.figure(figsize = (5,5))
    plt.pie(v, labels = [str(i) + "%" for i in vP], explode = ex, labeldistance = 1.3)
    plt.legend([i + " - " + str(j) for i, j in zip(l, v)], bbox_to_anchor=(2, 0.8))
    plt.title(n + " Categorization", pad = 15)
    plt.show()
```
