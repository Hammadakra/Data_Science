```python
from string import punctuation
def clean(s):
    s = s.lower()
    st = ""
    for c in s:
        if not (c.isnumeric() or c in punctuation):
            st += c
    return nltk.word_tokenize(st)
```

```python
senClean = clean(df.Reviews.iloc[0])
```

```python
words = []
for word in senClean:
    if word not in sw:
        words.append(word)
```

```python
def TC(st):# Term count
    senClean = clean(st)
    words = []
    for word in senClean:
        if word not in sw:
            words.append(word)
    return Counter(words)
```

```python
TCDF = df.Reviews.apply(TC)
```

```python
def TF(ctr):
    nofwords = sum(ctr.values())
    for item in ctr.items():
        key = item[0]
        value = item[1]
        ctr[key] = value / nofwords
    return ctr
```

```python
TCDF = TCDF.apply(TF)
```
