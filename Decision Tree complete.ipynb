```python
sliceing=df.iloc[:,:-1]
sliceing.head()
```

```python
test_y=df.Target[600:]
test_y.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics 
```

```python
DecTreeObj = DecisionTreeClassifier(criterion='entropy')

DecTreeObj = DecTreeObj.fit(train_x,train_y)

pred_y = DecTreeObj.predict(test_x)
```

```python
print("Accuracy:",metrics.accuracy_score(test_y, pred_y))
```

# With Out Shuffling

```python
import pandas as pd
from sklearn.utils import shuffle 
```

```python
data=pd.read_csv("daily_weather.csv")
data
```

```python
data['Target']=[ 1 if data.relative_humidity_3pm[i] > 40 and data.relative_humidity_9am[i] > 40 else 0 for i in range(len(data.relative_humidity_3pm))]
```

```python
data
```

```python
data.dropna(axis=0, inplace = True)
data.isna().sum()
```

```python
sliceing=data.iloc[:,:-1]
sliceing.head()



train_x=sliceing[:600]

train_y=data.Target[:600]
train_y.head()

test_x=sliceing[600:]
test_x.head()

test_y=data.Target[600:]
test_y.head()
```

```python
test_x.values
```

```python

 

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics 

DecTree = DecisionTreeClassifier(criterion='entropy')

DecTree= DecTree.fit(train_x,train_y)

pred_y = DecTree.predict(test_x)

print("Accuracy:",metrics.accuracy_score(test_y, pred_y))
```
